```python
import matplotlib.pyplot as plt
```

```python

# Example plot (e.g., a bar plot)
plt.bar([1, 2, 3, 4], [10, 15, 7, 20])

# Add y-gridlines
plt.grid(axis='y', linestyle='--', color='mediumslateblue', alpha=0.4)  # Customize as needed

# Add title and labels
plt.title('Bar Plot with Y-Gridlines', fontsize=16, fontweight='bold')
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

# Display the plot
plt.show()
```
